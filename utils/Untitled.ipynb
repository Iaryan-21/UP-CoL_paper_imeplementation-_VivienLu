```python
import csv
import os

log_file_path = '/home/amishr17/aryan/aorta-dissec/opt_folder/results/TBAD/snapshot/snapshot.log'
csv_file_path = '/home/amishr17/aryan/aorta-dissec/opt_folder/results/snapshot.csv'

# Headers need to exactly match the keys in the data part of your log lines
headers = ['epoch', 'step', 'loss_all', 'loss_ce', 'loss_dice', 'loss_focal', 'loss_iou', 'loss_supervised', 'loss_pc_lab', 'loss_pc_unlab', 'train_dice']

with open(log_file_path, 'r') as log_file, open(csv_file_path, 'w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=headers)
    writer.writeheader()
    
    for line in log_file:
        # Find the starting point of the data by locating "INFO" and adjusting the start index accordingly
        data_start_index = line.find("INFO") + len("INFO")
        if data_start_index > len("INFO"):  # Ensure "INFO" was found
            data_part = line[data_start_index:]
            try:
                # Now proceed with extracting key-value pairs from the data part
                data = {}
                for part in data_part.split(','):
                    key_value = part.split(':')
                    if len(key_value) == 2:
                        key, value = key_value[0].strip(), key_value[1].strip()
                        if key in headers:  # Make sure the key is one of the headers
                            data[key] = value
                
                # Check if all expected data was found before writing
                if all(header in data for header in headers):
                    writer.writerow(data)
                else:
                    print(f"Missing data for headers in line: {line.strip()}")
            except Exception as e:
                print(f"Error processing line: {line.strip()}\nError: {e}")
        else:
            print(f"Skipping line due to missing 'INFO' marker: {line.strip()}")

print("Conversion complete. CSV saved at", csv_file_path)
```
